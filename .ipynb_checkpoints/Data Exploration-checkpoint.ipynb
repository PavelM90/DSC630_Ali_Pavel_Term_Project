## Data exploration


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Salaries data frame
df_salaries = pd.read_csv("salaries.csv")
```

```python
# Get information
df_salaries.info()
```

```python
# Check the data frame
df_salaries.head()
```

```python
# Get quick stats
df_salaries.describe()
```

```python
# Donwload survey data frame
df_survey = pd.read_csv("salary_data.csv")
```

```python
# Check the data frame
df_survey.head()
```

```python
# Convert to pandas data frame
df_survey = pd.DataFrame(data = df_survey)
```

```python
# Check the data frame
df_survey.head()
```

```python
# Strip the data frame by ";" delimeter
df_survey[['ID', 'Income', 'Age', 'Gender', 'Education_Level']] = df_survey['ID;income;age;gender;education_level'].str.split(';', expand=True)
```

```python
# Check the df
df_survey.head()
```

```python
df_survey = pd.DataFrame(data = df_survey)
```

```python
df_survey["Income"] = pd.to_numeric(df_survey["Income"], errors='coerce')
df_survey["Age"] = pd.to_numeric(df_survey["Age"], errors='coerce')
```

```python
# Drop the combined column
df_survey.drop(columns= "ID;income;age;gender;education_level", inplace = True)
```

```python
# Save formatted csv 
df_survey.to_csv('survey_formated.csv')
```

```python
# Get the df info
df_survey.info()
```

```python
# Get quick stats
df_survey.describe()
```

```python
# Plot histogram for income distribution from salary data frame 
plt.figure()
plt.hist(df_salaries['salary_in_usd'], bins = 20)
plt.show()
```

```python
# Plot histogram for age from survey data
plt.figure()
plt.hist(df_survey['Age'], bins = 20)
plt.show()
```

```python
# Plot a histogram for income from survey data
plt.figure()
plt.hist(df_survey['Income'], bins = 45)
plt.show()
```
